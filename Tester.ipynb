```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
generator = tf.keras.models.load_model("Models/Generator100Epochs")
```

```python
def display_image(class_to_display):
    noise_vector = tf.random.normal(shape=(1, 32))
    class_vector = [class_to_display]
    class_vector = tf.one_hot(class_vector, depth=10)
    noise_vector = tf.concat([noise_vector, class_vector], axis = 1)
    fake_image = generator(noise_vector)
    fake_image = fake_image[0]
    plt.imshow(fake_image)
    plt.show()
    pass
```

```python
for i in range(10):
    display_image(i)
```
